```python
# How it works? Very simple
import string as str
letters = str.ascii_lowercase
def has_double_letter(letter):
    for word in scrabble.wordlist:
        if letter + letter in word:
            return True
    return False
# Print them       
for letter in letters:
    if not has_double_letter(letter):
        print(letter + " never appears doubled")
```

```python
# How many of these words have vowels
vowels = 'aeiou'

def has_vowels(word):
    for vowel in vowels:
        if vowel not in word:
            return False
    return True
counter = 0 # we set the counter
for word in scrabble.wordlist:
    if has_vowels(word):
        counter = counter + 1
        print(word)
print(counter) # how many words
```

```python
print(counter + ' words with vowels in them!')
```
